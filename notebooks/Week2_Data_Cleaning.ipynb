Cell 1 - Import Libraries

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
```

Cell 2 - Load Dataset

```python
branches = pd.read_csv("../data/raw/branches.csv")
customers = pd.read_csv("../data/raw/customers.csv")
inventory = pd.read_csv("../data/raw/inventory_master.csv")
invoices = pd.read_csv("../data/raw/invoices.csv")
payments = pd.read_csv("../data/raw/payments.csv")
products = pd.read_csv("../data/raw/products.csv")
po_header = pd.read_csv("../data/raw/purchase_orders_header.csv")
po_lines = pd.read_csv("../data/raw/purchase_orders_lines.csv")
sales_header = pd.read_csv("../data/raw/sales_orders_header.csv")
sales_lines = pd.read_csv("../data/raw/sales_orders_lines.csv")
stock = pd.read_csv("../data/raw/stock_ledger.csv")
suppliers = pd.read_csv("../data/raw/suppliers.csv")
```

Cell 3 - Store in Dictionary

```python
datasets = {
    "Branches": branches,
    "Customers": customers,
    "Inventory": inventory,
    "Invoices": invoices,
    "Payments": payments,
    "Products": products,
    "Purchase Header": po_header,
    "Purchase Lines": po_lines,
    "Sales Header": sales_header,
    "Sales Lines": sales_lines,
    "Stock Ledger": stock,
    "Suppliers": suppliers
}
```

Cell 4 - Standardize Column Names

```python
for name, df in datasets.items():

    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
    )

print("Column names standardized.")
```

Cell 5 - Remove Duplicate Rows

```python
duplicate_summary = []

for name, df in datasets.items():

    before = len(df)

    df.drop_duplicates(inplace=True)

    after = len(df)

    duplicate_summary.append({
        "Dataset": name,
        "Duplicates Removed": before - after
    })

pd.DataFrame(duplicate_summary)
```

Cell 6 - Missing Value Report

```python
missing_report = []

for name, df in datasets.items():

    missing_report.append({
        "Dataset": name,
        "Missing Values": df.isnull().sum().sum()
    })

pd.DataFrame(missing_report)
```

Cell 7 - Fill Missing Values

```python
for name, df in datasets.items():

    for col in df.columns:

        # Numeric columns
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col] = df[col].fillna(df[col].median())

        # Date columns
        elif pd.api.types.is_datetime64_any_dtype(df[col]):
            # Leave missing dates as NaT
            pass

        # Text / categorical columns
        else:
            df[col] = df[col].fillna("Unknown")

print("Missing values handled.")
```

Cell 8 - Remove Extra Spaces

```python
for name, df in datasets.items():

    text_cols = df.select_dtypes(include="object").columns

    for col in text_cols:

        df[col] = df[col].str.strip()

print("Whitespace removed.")
```

Cell 9 - Convert Text to Title Case

```python
for name, df in datasets.items():

    text_cols = df.select_dtypes(include="object").columns

    for col in text_cols:

        df[col] = df[col].str.title()

print("Text standardized.")
```

Cell 10 - Convert Date Columns

```python
for name, df in datasets.items():

    for col in df.columns:

        if "date" in col:

            df[col] = pd.to_datetime(
                df[col],
                errors="coerce"
            )

print("Date columns converted.")
```

Cell 11 - Dataset Quality Check

```python
quality = []

for name, df in datasets.items():

    quality.append({

        "Dataset": name,

        "Rows": len(df),

        "Columns": len(df.columns),

        "Missing": df.isnull().sum().sum(),

        "Duplicates": df.duplicated().sum()

    })

quality_df = pd.DataFrame(quality)

quality_df
```

Cell 12 - Invalid Numeric Values

```python
for name, df in datasets.items():

    numeric = df.select_dtypes(include=np.number)

    print("="*60)

    print(name)

    for col in numeric.columns:

        negative = (df[col] < 0).sum()

        print(col, ":", negative, "negative values")
```

Cell 13 - Feature Engineering

```python
for name, df in datasets.items():

    date_columns = [col for col in df.columns if "date" in col]

    for col in date_columns:

        df[col] = pd.to_datetime(df[col], errors="coerce")

        df[f"{col}_year"] = df[col].dt.year
        df[f"{col}_month"] = df[col].dt.month

print("Feature engineering completed.")
```

Cell 14 - Save Cleaned Dataset

```python
import os

os.makedirs("../data/cleaned", exist_ok=True)

for name, df in datasets.items():

    filename = name.lower().replace(" ", "_")

    df.to_csv(
        f"../data/cleaned/{filename}.csv",
        index=False
    )

print("Cleaned datasets saved.")
```

```python
summary = pd.DataFrame({

    "Dataset":[x for x in datasets.keys()],

    "Rows":[df.shape[0] for df in datasets.values()],

    "Columns":[df.shape[1] for df in datasets.values()]

})

summary
```
